```python
# altering the dataset
# inserts a new column named size which contains the column bedrooms multiplied by 20
dataset['size'] = (dataset['bedrooms']*20)
```

```python
dataset['size']
```

```python
# creating a function that categorizes by the size
def categorize(s):
    if s >= 80:
        return 'Big'
    elif s >= 60:
        return 'Medium'
    elif s >= 40:
        return 'Small'
```

```python
dataset['cat_size'] = dataset['size'].apply(categorize)
```

```python
dataset['cat_size']
```

```python
# Seing the column cat_size distribution
pd.value_counts(dataset['cat_size'])
```

```python
# axis=1 allows to exclude a column 
dataset.drop(['cat_size'], axis=1, inplace=True)
```

```python
dataset.drop(dataset[dataset.bedrooms>30].index, inplace=True)
```

```python
# Missing values
dataset.isnull().sum()
```

```python
# Removing the rows with missing values
dataset.dropna(inplace=True)
```

```python
dataset.isnull().sum()
```

```python
# it is possible to remove only rows with missing values in all columns
dataset.dropna(how='all', inplace=True)
```

```python
# fill the null values in the column floors with the mean
dataset['floors'].fillna(dataset['floors'].mean(), inplace=True)
```

```python
# fill the null values with 1 in the column bedrooms
dataset['bedrooms'].fillna(1, inplace=True)
```
